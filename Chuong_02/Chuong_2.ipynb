```python
import xml.dom.minidom

def main():
    doc = xml.dom.minidom.parse("sample.xml")

    staff_list = doc.getElementsByTagName("staff")

    print("Danh sách nhân viên:")
    for staff in staff_list:
        name = staff.getElementsByTagName("name")[0].firstChild.data
        salary = staff.getElementsByTagName("salary")[0].firstChild.data
        staff_id = staff.getAttribute("id")
        print(f"ID: {staff_id} | Tên: {name} | Lương: {salary}")

if __name__ == "__main__":
    main()
```

Bài 2.6:

```python
import requests
import xml.etree.ElementTree as ET
import csv

def main():
    url = "http://www.hindustantimes.com/rss/topnews/rssfeed.xml"
    response = requests.get(url)

    with open("rssfeed.xml", "wb") as file:
        file.write(response.content)
    print("Đã tải và lưu rssfeed.xml")

    tree = ET.parse("rssfeed.xml")
    root = tree.getroot()

    items = root.findall("./channel/item")

    news_data = []

    for item in items:
        title = item.find("title").text if item.find("title") is not None else ""
        link = item.find("link").text if item.find("link") is not None else ""
        pubDate = item.find("pubDate").text if item.find("pubDate") is not None else ""
        description = item.find("description").text if item.find("description") is not None else ""
        
        news_data.append([title, link, pubDate, description])

    print(f"Đã phân tích {len(news_data)} tin tức từ RSS feed")

    with open("news.csv", "w", newline="", encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Title", "Link", "Published Date", "Description"])
        writer.writerows(news_data)

    print("Đã lưu tin tức vào file news.csv")

if __name__ == "__main__":
    main()
```

Bài 2.10:

```python
import json
from collections import defaultdict 

FILE_NAME = 'Bai_2.10.json'

def thong_ke_nhan_vien():
    try:
        with open(FILE_NAME, 'r', encoding='utf-8') as f:
            data = json.load(f) 

    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp {FILE_NAME}. Hãy tạo tệp dữ liệu trước.")
        return
    except json.JSONDecodeError:
        print("Lỗi: Tệp JSON không hợp lệ.")
        return

    danh_sach_nhan_vien = data.get("Nhan_vien", [])
    tong_so_nhan_vien = len(danh_sach_nhan_vien)

    print(f"Tên công ty: {data.get('Ten_cong_ty', 'N/A')}")
    print(f"Địa chỉ: {data.get('Dia_chi', 'N/A')}")
    print(f"Tổng số nhân viên: {tong_so_nhan_vien}")
    print("-----Thống kê nhân viên theo đơn vị-----")

    dem_nhan_vien_theo_don_vi = defaultdict(int)
    for nv in danh_sach_nhan_vien:
        don_vi = nv.get("Don_vi")
        if don_vi:
            dem_nhan_vien_theo_don_vi[don_vi] += 1
    
    stt = 1
    for don_vi, so_luong in sorted(dem_nhan_vien_theo_don_vi.items()):
        ty_le = (so_luong / tong_so_nhan_vien) * 100
        
        print(f"{stt}./Tên đơn vị: {don_vi}")
        print(f"  - Số nhân viên: {so_luong}")
        print(f"  - Tỷ lệ: {ty_le:.2f}% so với tổng số nhân viên.") 
        stt += 1

if __name__ == "__main__":
    thong_ke_nhan_vien()
```

Bài 2.11:

```python
import json
import datetime
import sys 


danh_sach_giao_dich = []

def nhap_giao_dich_tu_nguoi_dung():

    while True:
        ma_giao_dich = input("Nhập Mã Giao Dịch (hoặc 'ketthuc' để dừng): ")
        if ma_giao_dich.lower() == 'ketthuc':
            break
            
        loai_tai_san = input("Loại tài sản (Tien/Vang): ")
        try:
            gia_tri = float(input("Giá trị giao dịch: "))
        except ValueError:
            print("Lỗi: Giá trị phải là số. Vui lòng nhập lại giao dịch này.")
            continue
            
        danh_sach_giao_dich.append({
            "ma_gd": ma_giao_dich,
            "tai_san": loai_tai_san,
            "gia_tri": gia_tri,
            "ngay_nhap": datetime.datetime.now().strftime("%Y-%m-%d")
        })
        print("-> Đã thêm giao dịch.")
    
    print(f"\nTổng số giao dịch đã nhập: {len(danh_sach_giao_dich)}")

def ghi_du_lieu_json(data):
    if not data:
        print("Không có giao dịch để ghi vào tệp.")
        return

    timestamp_format = "%Y-%m-%d-%H-%M-%S"
    ten_file = datetime.datetime.now().strftime(timestamp_format) + ".json"
    
    try:
        with open(ten_file, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4) 
        
        print(f"\nDữ liệu đã được ghi thành công vào tệp: {ten_file}")
        
    except IOError as e:
        print(f"\nLỗi khi thao tác với tệp: {e}")



if __name__ == "__main__":

    nhap_giao_dich_tu_nguoi_dung()

    if not danh_sach_giao_dich:
        print("Không có giao dịch nào được thực hiện. Kết thúc chương trình.")
        sys.exit()


    while True:
        lua_chon = input("\nBạn có muốn ghi dữ liệu giao dịch vào tập tin không? (Chọn 1 hoặc 0): ")
        
        if lua_chon == '1':
            ghi_du_lieu_json(danh_sach_giao_dich)
            break
        elif lua_chon == '0':
            print("Không ghi dữ liệu vào tập tin. Chương trình kết thúc.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng nhập 1 hoặc 0.")
```

Bài 2.12:

```python
import requests

API_URL = 'https://jsonplaceholder.typicode.com/posts'

def doc_va_hien_thi_bai_post():
    try:
        response = requests.get(API_URL)

        if response.status_code == 200:
            
            json_data = response.json()
            
            tong_so_bai_post = len(json_data)
            
            print(f"Tổng số bài post: {tong_so_bai_post}")
            
            print("Danh sách các bài post (userId, id, title, body):")
            
            so_bai_hien_thi = min(5, tong_so_bai_post) 

            for i in range(so_bai_hien_thi):
                post = json_data[i]
                
                print(f"Bài {i+1} (ID: {post['id']}):")
                print(f"- User ID: {post['userId']}")
                print(f"- Tiêu đề: {post['title']}")
                print(f"- Nội dung: {post['body'][:100]}...") 
                
            if tong_so_bai_post > so_bai_hien_thi:
                print(f"(... Chỉ hiển thị {so_bai_hien_thi} bài đầu tiên trong tổng số {tong_so_bai_post} bài post)")


        else:
            print(f"Yêu cầu không thành công. Mã trạng thái: {response.status_code}")

    except requests.exceptions.RequestException:
        print("Lỗi kết nối: Không thể truy cập API. Vui lòng kiểm tra mạng.")

if __name__ == "__main__":
    doc_va_hien_thi_bai_post()
```

```python
import requests

API_URL = 'https://jsonplaceholder.typicode.com/comments?postId=1'
GIOI_HAN_HIEN_THI = 3

def doc_va_hien_thi_sach_noi_bat():
    try:
        response = requests.get(API_URL)

        if response.status_code == 200:
            
            json_data = response.json()
            
            tong_so_bai = len(json_data)
            
            print(f"Tổng số 'sách nổi bật' (bài post/comment) có postId=1: {tong_so_bai}")
            print("----------------------------------------")
            
            print("Danh sách các bài post nổi bật (Giới hạn hiển thị 3 bài):")
            
            so_bai_hien_thi = min(GIOI_HAN_HIEN_THI, tong_so_bai)
            
            for i in range(so_bai_hien_thi):
                bai_post = json_data[i]
                
                print(f"Bài {i+1} (ID: {bai_post['id']}):")
                print(f"- Tên: {bai_post['name']}")
                print(f"- Email: {bai_post['email']}")
                print(f"- Nội dung: {bai_post['body'][:100]}") 
            
            if tong_so_bai > GIOI_HAN_HIEN_THI:
                print(f"(... và {tong_so_bai - GIOI_HAN_HIEN_THI} bài khác.)")

        else:
            print(f"Yêu cầu không thành công. Mã trạng thái: {response.status_code}")

    except requests.exceptions.RequestException:
        print("Lỗi kết nối: Không thể truy cập API. Vui lòng kiểm tra mạng.")

if __name__ == "__main__":
    doc_va_hien_thi_sach_noi_bat()
```
