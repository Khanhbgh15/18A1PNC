```python
import pandas as pd

ser1 = pd.Series([1,2,3,4,5])
ser2 = pd.Series([4,5,6,7,8])

res1 = ser1[~ser1.isin(ser2)].reset_index(drop=True)

sym_vals = list(set(ser1.tolist()).symmetric_difference(set(ser2.tolist())))
res2 = pd.Series(sym_vals).sort_values().reset_index(drop=True)

print("ser1 sau khi xóa các phần tử xuất hiện trong ser2:\n", res1)
print("\nGiá trị thuộc ser1 hoặc ser2 nhưng không cùng xuất hiện ở cả hai:\n", res2)
```

```python
euro12 = pd.read_csv('euro2012.csv')

#in type, shape, list cột
print(type(euro12))
print("Shape:", euro12.shape)
print("Columns:", euro12.columns.tolist())

#in giá trị cột Goals
print("\nGoals:\n", euro12['Goals'])

#có bao nhiêu đội tham gia Euro 2012?
print("\nSố đội:", euro12['Team'].nunique())

#in thông tin Euro 2012
print("\nThông tin Euro 2012:\n")
print(euro12.info())

#tạo dataframe discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\ndiscipline:\n", discipline)

#sắp xếp giảm dần theo Red Cards, Yellow Cards
discipline_sorted = discipline.sort_values(
    by=['Red Cards', 'Yellow Cards'], ascending=False)
print("\nSắp xếp discipline:\n", discipline_sorted)

#tính trung bình Yellow Cards
avg_yellow = euro12['Yellow Cards'].mean()
print("\nTrung bình Yellow Cards:", avg_yellow)

#lọc các đội ghi hơn 6 bàn
more_than_6_goals = euro12[euro12['Goals'] > 6]
print("\nĐội ghi > 6 Goals:\n", more_than_6_goals)

#in các đội bắt đầu bằng 'G'
teams_start_G = euro12[euro12['Team'].str.startswith('G')]
print("\n7. Team bắt đầu bằng G:\n", teams_start_G)

#in 7 cột đầu
print("\n7 cột đầu:\n", euro12.iloc[:, :7])

#in tất cả trừ 3 cột cuối
print("\nTrừ 3 cột cuối:\n", euro12.iloc[:, :-3])

#in Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
print("\nCột được yêu cầu:\n",
      euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']])

#chỉ hiển thị England, Italy, Russia
filtered3 = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])]
print("\nTeam + Shooting Accuracy:\n", 
      filtered3[['Team', 'Shooting Accuracy']])
```

```python
import pandas as pd

data = {
    'country': ['Vietnam', 'Japan', 'China', 'France', 'Germany', 'USA', 'Canada', 'Brazil', 'India', 'Russia'],
    'continent': ['Asia', 'Asia', 'Asia', 'Europe', 'Europe', 'North America', 'North America', 'South America', 'Asia', 'Europe'],
    'beer_servings': [111, 77, 45, 127, 346, 249, 240, 245, 10, 247],
    'spirit_servings': [2, 16, 22, 151, 117, 158, 122, 145, 3, 326],
    'wine_servings': [1, 14, 8, 370, 275, 84, 100, 16, 4, 73]
}

drinks_df = pd.DataFrame(data)

drinks_df.to_csv('drinks.csv', index=False, encoding='utf-8')
print("Đã tạo file drinks.csv thành công!")
```

```python
drink = pd.read_csv('drinks.csv', index_col=0)

print(type(drink))
print("Shape:", drink.shape)
print("Columns:", drink.columns.tolist())
print("\nHead:\n", drink.head())
print("\nTail:\n", drink.tail())

#trung bình bia mỗi châu lục
print("\nTrung bình bia theo châu lục:\n",
      drink.groupby('continent')['beer_servings'].mean())

#thống kê rượu vang theo châu lục
print("\nThống kê wine_servings theo continent:\n",
      drink.groupby('continent')['wine_servings'].describe())

#trung bình bia và rượu mỗi châu lục
print("\nMean bia + rượu theo châu lục:\n",
      drink.groupby('continent')[['beer_servings',
                                  'wine_servings']].mean())

#median bia + rượu mỗi châu lục
print("\nMedian bia + rượu theo châu lục:\n",
      drink.groupby('continent')[['beer_servings',
                                  'wine_servings']].median())

#spirit_servings trung bình, max, min theo châu lục
print("\nSpirit stats theo continent:\n",
      drink.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min']))

#sort theo beer_servings
sorted_beer = drink.sort_values(by='beer_servings')

print("\n5 nước uống nhiều bia nhất:\n",
      sorted_beer.tail(5))

print("\n5 nước uống ít bia nhất:\n",
      sorted_beer.head(5))
```

```python
import pandas as pd
import numpy as np

stocks1_data = {
    'date': ['2024-01-01', '2024-01-01', '2024-01-02'],
    'symbol': ['AAPL', 'MSFT', 'AAPL'],
    'open': [150, 330, 152],
    'high': [155, np.nan, 158],
    'low': [148, 325, np.nan],         
    'close': [154, 329, 155],
    'volume': [80000, 65000, 90000]
}
pd.DataFrame(stocks1_data).to_csv("stocks1.csv", index=False)

stocks2_data = {
    'date': ['2024-01-03', '2024-01-03'],
    'symbol': ['AAPL', 'MSFT'],
    'open': [153, 331],
    'high': [159, 335],
    'low': [150, 327],
    'close': [158, 333],
    'volume': [85000, 67000]
}
pd.DataFrame(stocks2_data).to_csv("stocks2.csv", index=False)


companies_data = {
    'name': ['AAPL', 'MSFT'],
    'employees': [164000, 221000],
    'headquarters_city': ['Cupertino', 'Redmond'],
    'headquarters_state': ['California', 'Washington']
}
pd.DataFrame(companies_data).to_csv("companies.csv", index=False)
```

```python
stocks1 = pd.read_csv("stocks1.csv")
stocks2 = pd.read_csv("stocks2.csv")
companies = pd.read_csv("companies.csv")


print(stocks1.head())
print(stocks1.tail())
print(stocks1.dtypes)
print(stocks1.info())


print(stocks2.head())
print(stocks2.tail())
print(stocks2.dtypes)
print(stocks2.info())

print(companies)
print(companies.dtypes)
print(companies.info())
```

```python
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(
    lambda x: x.fillna(x.max())
)

stocks1['low'] = stocks1.groupby('symbol')['low'].transform(
    lambda x: x.fillna(x.min())
)

print(stocks1)
```

```python
stocks = pd.concat([stocks1, stocks2], ignore_index=True)
print(stocks.tail(15))
```

```python
stocks_companies = stocks.merge(companies, left_on="symbol", right_on="name", how="left")
print(stocks_companies.head())
```

```python
print(stocks.groupby("symbol")[['open','high','low','close','volume']].mean())
```

```python
print(stocks.groupby("symbol")['close'].agg(['mean', 'max', 'min']))
```

```python
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print(stocks_companies['parsed_time'].dtype)
print(stocks_companies.head())
```

```python
stocks_companies['result'] = stocks_companies.apply(
    lambda row: 'up' if row['close'] > row['open'] else 'down',
    axis=1
)
print(stocks_companies[['symbol','open','close','result']])
```
