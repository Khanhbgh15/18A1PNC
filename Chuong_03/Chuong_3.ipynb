# Bài 3.5

```python
import numpy as np

np.random.seed(0)
arr_height = np.random.normal(70, 3, 200)  
arr_weight = np.random.normal(180, 20, 200) 
```

```python
#kiểm tra kích thước dữ liệu
print("Kích thước arr_height:", arr_height.shape)
print("Kích thước arr_weight:", arr_weight.shape)
```

```python
#quy đổi inch - mét (1 inch = 0.0254 m)
he_so_quy_doi_chieu_cao = 0.0254
arr_height_m = arr_height * he_so_quy_doi_chieu_cao
print("Chiều cao (m):", arr_height_m[:5]) 
```

```python
#quy đổi pound - kg (1 pound = 0.453592 kg)
he_so_quy_doi_can_nang = 0.453592
arr_weight_kg = arr_weight * he_so_quy_doi_can_nang
print("Cân nặng (kg):", arr_weight_kg[:5])
```

```python
#tính BMI = cân nặng / (chiều cao^2)
arr_bmi = arr_weight_kg / (arr_height_m ** 2)
print("Chỉ số BMI (5 người đầu tiên):", arr_bmi[:5])
```

```python
#giá trị cân nặng ở vị trí index = 50
print("Cân nặng cầu thủ thứ 51 (index 50):", arr_weight_kg[50])
```

```python
#lấy chiều cao từ index 100 đến 110 (bao gồm 110)
arr_height_m_100 = arr_height_m[100:111]
print("Chiều cao từ index 100 đến 110:", arr_height_m_100)
```

```python
#lọc cầu thủ có BMI < 21
bmi_duoi_21 = arr_bmi[arr_bmi < 21]
print("Số cầu thủ có BMI < 21:", len(bmi_duoi_21))
print("Các giá trị BMI < 21:", bmi_duoi_21)
```

```python
#chiều cao và cân nặng trung bình
chieu_cao_tb = np.mean(arr_height_m)
can_nang_tb = np.mean(arr_weight_kg)
print("\nChiều cao trung bình: {:.2f} m".format(chieu_cao_tb))
print("Cân nặng trung bình: {:.2f} kg".format(can_nang_tb))
```

```python
#chiều cao và cân nặng lớn nhất
chieu_cao_max = np.max(arr_height_m)
can_nang_max = np.max(arr_weight_kg)
print("Chiều cao lớn nhất: {:.2f} m".format(chieu_cao_max))
print("Cân nặng lớn nhất: {:.2f} kg".format(can_nang_max))
```

```python
#chiều cao và cân nặng nhỏ nhất
chieu_cao_min = np.min(arr_height_m)
can_nang_min = np.min(arr_weight_kg)
print("Chiều cao nhỏ nhất: {:.2f} m".format(chieu_cao_min))
print("Cân nặng nhỏ nhất: {:.2f} kg".format(can_nang_min))
```

```python
import numpy as np

baseball = [
    [74, 180],
    [72, 215],
    [73, 210],
    [69, 185],
    [70, 190],
    [76, 215],
    [71, 200],
    [73, 188],
    [75, 220],
    [77, 230]
]
```

```python
#tạo NumPy array từ list
np_baseball = np.array(baseball)
print("Mảng np_baseball:", np_baseball)
```

```python
# kiểu dữ liệu và kích thước
print("Kiểu dữ liệu của np_baseball:", np_baseball.dtype)
print("Kích thước mảng (shape):", np_baseball.shape)
```

```python
# in các giá trị dòng thứ 50
print("Dòng thứ 5 (minh họa cho dòng 50):", np_baseball[4])
```

```python
#tạo np_weight từ cột 2 (cột cân nặng)
np_weight = np_baseball[:, 1]
print("Mảng np_weight (cột cân nặng):", np_weight)
```

```python
#chiều cao của vận động viên thứ 124
print("Chiều cao vận động viên thứ 4 (minh họa):", np_baseball[3, 0], "inch")
```

```python
#tính chiều cao & cân nặng trung bình
chieu_cao_tb = np.mean(np_baseball[:, 0])
can_nang_tb = np.mean(np_baseball[:, 1])
print("Chiều cao trung bình: {:.2f} inch".format(chieu_cao_tb))
print("Cân nặng trung bình: {:.2f} pounds".format(can_nang_tb))
```

```python
#phân tích mối tương quan giữa chiều cao, cân nặng
corr = np.corrcoef(np_baseball[:, 0], np_baseball[:, 1])[0, 1]
print("Hệ số tương quan giữa chiều cao và cân nặng:", round(corr, 3))

if corr > 0.5:
    print("Có tương quan thuận (chiều cao tăng - cân nặng tăng).")
elif corr < -0.5:
    print("Có tương quan nghịch (chiều cao tăng - cân nặng giảm).")
else:
    print("Không có tương quan rõ ràng giữa chiều cao và cân nặng.")
```

# Bài 3.8

```python
import numpy as np

heights = [183, 175, 190, 178, 181, 188, 177, 172, 195, 180]
positions = ['GK', 'D', 'A', 'M', 'D', 'A', 'GK', 'M', 'D', 'A']
```

```python
#tạo numpy array từ list

# a) Tạo np_positions
np_positions = np.array(positions)
print("Mảng np_positions:", np_positions)
print("Kiểu dữ liệu:", np_positions.dtype)

# b) Tạo np_heights
np_heights = np.array(heights)
print("Mảng np_heights:", np_heights)
print("Kiểu dữ liệu:", np_heights.dtype)
```

```python
#tính chiều cao trung bình của các GK
mask_gk = np_positions == 'GK'
avg_gk = np.mean(np_heights[mask_gk])
print("Chiều cao trung bình của GK:", round(avg_gk, 2), "cm")
```

```python
#tính chiều cao trung bình của những vị trí KHÔNG phải GK
mask_not_gk = np_positions != 'GK'
avg_not_gk = np.mean(np_heights[mask_not_gk])
print("Chiều cao trung bình của các vị trí khác GK:", round(avg_not_gk, 2), "cm")
```

```python
#tạo mảng dữ liệu có cấu trúc
players_dtype = np.dtype([('position', 'U5'), ('height', 'f8')])
players = np.array(list(zip(np_positions, np_heights)), dtype=players_dtype)
print("Mảng structured array players:", players)
```

```python
#sắp xếp mảng players theo chiều cao
players_sorted = np.sort(players, order='height')
print("Mảng players sau khi sắp xếp theo chiều cao:", players_sorted)
```

```python
#cầu thủ cao nhất và thấp nhất
tallest = players_sorted[-1]
shortest = players_sorted[0]
print("Vị trí có chiều cao cao nhất: {} ({:.2f} cm)".format(tallest['position'], tallest['height']))
print("Vị trí có chiều cao thấp nhất: {} ({:.2f} cm)".format(shortest['position'], shortest['height']))
```

```python
import pandas as pd
import numpy as np

data_stocks1 = {
    'date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03'],
    'symbol': ['AAPL', 'GOOG', 'AAPL', 'GOOG', 'AAPL'],
    'open': [170.0, np.nan, 172.0, 96.0, 175.0], 
    'high': [171.0, 95.5, 172.5, 96.5, np.nan],
    'low': [169.5, 94.8, 171.5, 95.8, 174.5],
    'close': [170.5, 95.2, 172.2, 96.2, 175.2],
    'volume': [100000, 200000, 120000, 210000, 150000]
}

data_stocks2 = {
    'date': ['2023-01-03', '2023-01-04', '2023-01-04'],
    'symbol': ['GOOG', 'AAPL', 'GOOG'],
    'open': [97.0, 176.0, 98.0],
    'high': [97.5, 176.5, 98.5],
    'low': [96.8, 175.5, 97.8],
    'close': [97.2, 176.2, 98.2],
    'volume': [220000, 160000, 230000]
}

data_companies = {
    'symbol': ['AAPL', 'GOOG', 'MSFT'],
    'name': ['Apple Inc.', 'Alphabet Inc.', 'Microsoft Corp.'],
    'employees': [164000, 186000, 221000],
    'headquarters_city': ['Cupertino', 'Mountain View', 'Redmond'],
    'headquarters_state': ['CA', 'CA', 'WA']
}

stocks1 = pd.DataFrame(data_stocks1)
stocks2 = pd.DataFrame(data_stocks2)
companies = pd.DataFrame(data_companies)

stocks1['date'] = stocks1['date'].astype(str)
stocks2['date'] = stocks2['date'].astype(str)
```

```python
stocks1.head()
stocks1.dtypes
stocks1.info()

stocks2.head()
stocks2.dtypes
stocks2.info()

companies
companies.dtypes
companies.info()
```

```python
# Kiểm tra dữ liệu Null (NaN)
stocks1.isnull().sum()

max_high_by_symbol = stocks1.groupby('symbol')['high'].transform('max')
min_low_by_symbol = stocks1.groupby('symbol')['low'].transform('min')

stocks1['open'] = stocks1['open'].fillna(max_high_by_symbol)

stocks1['low'] = stocks1['low'].fillna(min_low_by_symbol)

stocks1
```

```python
# Gộp stocks1 và stocks2 theo chiều dọc 
stocks = pd.concat([stocks1, stocks2], ignore_index=True)

# Xem 15 dòng dữ liệu cuối của stocks
stocks.tail(15)
```

```python
# Gộp stocks và companies dựa trên cột 
stocks_companies = pd.merge(stocks, companies, on='symbol', how='left')

# Xem 5 dòng dữ liệu đầu của stocks_companies
stocks_companies.head()
```

```python
# Chọn các cột cần tính toán
cols_to_avg = ['open', 'high', 'low', 'close', 'volume']

# Group theo 'symbol' và tính trung bình
company_averages = stocks_companies.groupby('symbol')[cols_to_avg].mean()
company_averages
```

```python
# Group theo 'symbol' và sử dụng agg() để tính mean, max, min trên cột 'close'
close_stats = stocks_companies.groupby('symbol')['close'].agg(['mean', 'max', 'min'])
close_stats
```

```python
# Chuyển đổi cột 'date' sang kiểu DateTime
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])

# Cho biết kiểu dữ liệu của cột parsed_time
print(stocks_companies['parsed_time'].dtype) 

# Hiển thị 5 dòng dữ liệu đầu của stocks_companies
stocks_companies.head()
```

```python
# Áp dụng Boolean Masking và np.where để gán giá trị theo điều kiện:
stocks_companies['result'] = np.where(
    stocks_companies['close'] > stocks_companies['open'],
    'up',
    'down'
)

print(stocks_companies[['symbol', 'open', 'close', 'result']])
```

```python
#XỬ LÝ DL

# Thêm một dòng có rating thiếu để test
ratings.loc[len(ratings)] = [5, 6, np.nan, 1425943100]

# Tính trung bình rating cho việc điền giá trị
avg_rating = ratings['rating'].mean()

# Điền giá trị thiếu (fillna) bằng rating trung bình
ratings['rating'].fillna(avg_rating, inplace=True)

print("Ratings tail (sau khi điền NaN):", ratings.tail(2))
```

```python
#GỘP DATAFRAME

# Gộp ratings và movies trên cột 'movieId'
movie_data = pd.merge(ratings, movies, on='movieId', how='left')
print(movie_data.head())
```

```python
#LỌC DỮ LIỆU THEO YÊU CẦU

# Lọc phim có rating >= 4.5
high_rated_movies = movie_data[movie_data['rating'] >= 4.5]
print(high_rated_movies[['title', 'rating']])
```

```python
#THỐNG KÊ DỮ LIỆU

# Tính rating trung bình cho mỗi phim
avg_rating_per_movie = movie_data.groupby('title')['rating'].mean()
print(avg_rating_per_movie.sort_values(ascending=False).head())
```

```python
#PARSING TIMESTAMPS 

# Chuyển đổi cột 'timestamp' sang định dạng DateTime
movie_data['datetime'] = pd.to_datetime(movie_data['timestamp'], unit='s')
print("Kiểu dữ liệu mới:", movie_data['datetime'].dtype)
print("Kết quả:", movie_data[['timestamp', 'datetime']].head())
```
