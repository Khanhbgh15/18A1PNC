```python
import threading

def print_thread_name():
    print("Đang chạy luồng:", threading.current_thread().name)

threads = []

for i in range(5):
    t = threading.Thread(target=print_thread_name, name=f"Thread-{i+1}")
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Kết thúc chương trình")
```

Bài 6.2

```python
import threading
import time

def download_file(filename):
    print(f"Bắt đầu tải {filename}")
    time.sleep(2)  
    print(f"Hoàn thành tải {filename}")

files = ["file1.zip", "file2.zip", "file3.zip"]
threads = []

for f in files:
    t = threading.Thread(target=download_file, args=(f,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Tải xong tất cả file")
```

Bài 6.3

```python
import threading

def print_even():
    for i in range(30, 51):
        if i % 2 == 0:
            print("Chẵn:", i)

def print_odd():
    for i in range(30, 51):
        if i % 2 != 0:
            print("Lẻ:", i)

t1 = threading.Thread(target=print_even)
t2 = threading.Thread(target=print_odd)

t1.start()
t2.start()

t1.join()
t2.join()
```

Bài 6.4

```python
import threading

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    print(f"Giai thừa của {n} là {result}")

numbers = [5, 6, 7]

threads = []
for n in numbers:
    t = threading.Thread(target=factorial, args=(n,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

Bài 6.7

```python
import threading
import time

def web_task(name, delay, repeat):
    print(f"Starting {name}")
    for _ in range(repeat):
        print(f"{name}: {time.ctime()}")
        time.sleep(delay)
    print(f"Exiting {name}")

print("Starting Main Thread")

t1 = threading.Thread(target=web_task, args=("Google", 1, 4))
t2 = threading.Thread(target=web_task, args=("Yahoo", 2, 3))
t3 = threading.Thread(target=web_task, args=("Facebook", 3, 3))

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```

Bài 6.8

```python
import threading
import random

n = 40
lst = [random.randint(0, 10) for _ in range(n)]
print("List:", lst)

num_threads = 3
chunk_size = n // num_threads
total_sum = 0
lock = threading.Lock()

def partial_sum(sublist, idx):
    global total_sum
    print(f"Thread {idx}: ", sublist)
    s = sum(sublist)
    lock.acquire()
    total_sum += s
    lock.release()

threads = []

for i in range(num_threads):
    start = i * chunk_size
    end = None if i == num_threads - 1 else (i + 1) * chunk_size
    t = threading.Thread(
        target=partial_sum,
        args=(lst[start:end], i + 1)
    )
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Tổng list =", total_sum)
```

Bài 6.10

```python
import threading
import random

n = 30
lst = [random.randint(0, 100) for _ in range(n)]
print("List:", lst)

num_threads = 3
chunk_size = n // num_threads
max_values = []

def find_max(sublist):
    max_values.append(max(sublist))

threads = []

for i in range(num_threads):
    start = i * chunk_size
    end = None if i == num_threads - 1 else (i + 1) * chunk_size
    t = threading.Thread(target=find_max, args=(lst[start:end],))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Giá trị lớn nhất trong list:", max(max_values))
```
