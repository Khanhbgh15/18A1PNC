```python
import tkinter as tk

def submit():
    print("Student Name:", name_entry.get())
    print("Student ID:", id_entry.get())
    print("Password:", pass_entry.get())

window = tk.Tk()
window.title("Student Login")

tk.Label(window, text="Student Name:").grid(row=0, column=0, padx=5, pady=5)
tk.Label(window, text="Student ID:").grid(row=1, column=0, padx=5, pady=5)
tk.Label(window, text="Password:").grid(row=2, column=0, padx=5, pady=5)

name_entry = tk.Entry(window)
id_entry = tk.Entry(window)
pass_entry = tk.Entry(window, show="*")

name_entry.grid(row=0, column=1)
id_entry.grid(row=1, column=1)
pass_entry.grid(row=2, column=1)

tk.Button(window, text="Submit", command=submit).grid(row=3, column=1, pady=10)

window.mainloop()
```

```python
import tkinter as tk

def reverse_text():
    s = entry.get()
    result = ""
    for i in range(len(s)-1, -1, -1):
        result += s[i]
    label_result.config(text=result)

window = tk.Tk()
window.title("Reverse String")

tk.Label(window, text="Enter a word:").pack()
entry = tk.Entry(window, width=30)
entry.pack()

tk.Button(window, text="Validate", command=reverse_text).pack()
label_result = tk.Label(window, text="")
label_result.pack()

window.mainloop()
```

```python
#Liệt kê số ước
n = int(input("Enter N: "))
print("Divisors:", end=" ")
for i in range(1, n+1):
    if n % i == 0:
        print(i, end=" ")

#Gui liệt kê ước số
import tkinter as tk

def find_divisors():
    n = int(entry.get())
    res = ""
    for i in range(1, n+1):
        if n % i == 0:
            res += str(i) + " "
    result.config(text=res)

window = tk.Tk()
window.title("Divisors")

tk.Label(window, text="Enter the value of N").pack()
entry = tk.Entry(window)
entry.pack()

tk.Button(window, text="Validate", command=find_divisors).pack()
result = tk.Label(window, text="")
result.pack()

window.mainloop()
```

```python
import tkinter as tk
from math import gcd

def calculate():
    a = int(e1.get())
    b = int(e2.get())
    if choice.get() == "GCD":
        res = gcd(a, b)
        result.config(text=f"GCD({a},{b}) = {res}")
    else:
        res = a * b // gcd(a, b)
        result.config(text=f"LCM({a},{b}) = {res}")

window = tk.Tk()
window.title("GCD & LCM")

tk.Label(window, text="Enter value of m:").grid(row=0, column=0)
tk.Label(window, text="Enter value of n:").grid(row=1, column=0)

e1 = tk.Entry(window)
e2 = tk.Entry(window)
e1.grid(row=0, column=1)
e2.grid(row=1, column=1)

choice = tk.StringVar(value="GCD")
tk.OptionMenu(window, choice, "GCD", "LCM").grid(row=2, column=1)

tk.Button(window, text="Calculate", command=calculate).grid(row=3, column=1)
result = tk.Label(window, text="")
result.grid(row=4, column=1)

window.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show_info():
    name = name_entry.get()
    age = int(age_entry.get())
    if age >= 18:
        message = f"Xin chào {name}\nBạn đã trưởng thành!"
    else:
        message = f"Xin chào {name}\nBạn còn nhỏ tuổi!"
    messagebox.showinfo("Thông báo", message)

window = tk.Tk()
window.title("User Info")

tk.Label(window, text="Name:").grid(row=0, column=0)
tk.Label(window, text="Age:").grid(row=1, column=0)

name_entry = tk.Entry(window)
age_entry = tk.Entry(window)
name_entry.grid(row=0, column=1)
age_entry.grid(row=1, column=1)

tk.Button(window, text="Submit", command=show_info).grid(row=2, column=1)

window.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

can = ["Giáp","Ất","Bính","Đinh","Mậu","Kỷ","Canh","Tân","Nhâm","Quý"]
chi = ["Tý","Sửu","Dần","Mão","Thìn","Tỵ","Ngọ","Mùi","Thân","Dậu","Tuất","Hợi"]

def convert():
    year = int(entry.get())
    name = can[year % 10] + " " + chi[year % 12]
    messagebox.showinfo("Kết quả", f"Năm âm lịch: {name}")

window = tk.Tk()
window.title("Can Chi")

tk.Label(window, text="Enter year:").pack()
entry = tk.Entry(window)
entry.pack()
tk.Button(window, text="Convert", command=convert).pack()

window.mainloop()
```

```python
import tkinter as tk

def calc_bmi():
    w = float(weight.get())
    h = float(height.get())
    bmi = w / (h*h)
    result.config(text=f"BMI: {bmi:.2f}")

window = tk.Tk()
window.title("BMI Calculator")

tk.Label(window, text="Weight (kg):").pack()
weight = tk.Entry(window)
weight.pack()

tk.Label(window, text="Height (m):").pack()
height = tk.Entry(window)
height.pack()

tk.Button(window, text="Calculate BMI", command=calc_bmi).pack()
result = tk.Label(window, text="")
result.pack()

window.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show():
    messagebox.showinfo("Lập trình hướng đối tượng",
                        "Giải phương trình bậc 2")

window = tk.Tk()
window.title("Hệ thống bài tập nâng cao Python")

menu = tk.Menu(window)
window.config(menu=menu)

chap1 = tk.Menu(menu, tearoff=0)
menu.add_cascade(label="Chương 1", menu=chap1)
chap1.add_command(label="1.1", command=show)
chap1.add_separator()
chap1.add_command(label="Exit", command=window.quit)

window.mainloop()
```

```python
import tkinter as tk

def calc_fare():
    km = float(entry.get())
    fare = 12000
    if km > 0.8:
        fare += (km - 0.8) * 15000
    result.config(text=f"Taxi fare: {fare:.0f} VND")

window = tk.Tk()
window.title("Taxi Fare")

tk.Label(window, text="Enter distance (km):").pack()
entry = tk.Entry(window)
entry.pack()

tk.Button(window, text="Calculate", command=calc_fare).pack()
result = tk.Label(window, text="")
result.pack()

window.mainloop()
```
