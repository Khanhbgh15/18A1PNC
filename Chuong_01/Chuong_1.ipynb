Bài 1

```python
class HinhChuNhat:
    def __init__(self, dai, rong):
        self.dai = dai
        self.rong = rong

    def chu_vi(self):
        return 2 * (self.dai + self.rong)

    def dien_tich(self):
        return self.dai * self.rong

    def hien_thi(self):
        print(f"Dài: {self.dai}, Rộng: {self.rong}")
        print(f"Chu vi: {self.chu_vi()}")
        print(f"Diện tích: {self.dien_tich()}")


a = float(input("Nhập chiều dài: "))
b = float(input("Nhập chiều rộng: "))
hcn = HinhChuNhat(a, b)
hcn.hien_thi()
```

Bài 2

```python
class ThiSinh:
    def __init__(self, hoten, toan, ly, hoa):
        self.hoten = hoten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def hien_thi(self):
        print(self.hoten, self.toan, self.ly, self.hoa, self.tong_diem())


ds = []
n = int(input("Nhập số thí sinh: "))
for _ in range(n):
    ht = input("Họ tên: ")
    t = float(input("Toán: "))
    l = float(input("Lý: "))
    h = float(input("Hóa: "))
    ds.append(ThiSinh(ht, t, l, h))

print("Danh sách thí sinh trúng tuyển:")
for ts in ds:
    if ts.tong_diem() >= 20:
        ts.hien_thi()
```

```python
import math

class PhanSo:
    def __init__(self, tu, mau):
        self.tu = tu
        self.mau = mau

    def hop_le(self):
        return self.mau != 0

    def rut_gon(self):
        ucln = math.gcd(self.tu, self.mau)
        self.tu //= ucln
        self.mau //= ucln

    def hien_thi(self):
        if self.hop_le():
            self.rut_gon()
            print(f"{self.tu}/{self.mau}")
        else:
            print("Phân số không hợp lệ")
```

Bài 4,5,6

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.data = []

    def push(self, x):
        if not self.isFull():
            self.data.append(x)

    def pop(self):
        if not self.isEmpty():
            return self.data.pop()

    def isEmpty(self):
        return len(self.data) == 0

    def isFull(self):
        return len(self.data) == self.size

    def count(self):
        return len(self.data)

    def print(self):
        print(self.data)
```

```python
from datetime import date, timedelta

class Date:
    def __init__(self, d=1, m=1, y=2000):
        self.ngay = date(y, m, d)

    def display(self):
        print(self.ngay)

    def next(self):
        self.ngay += timedelta(days=1)
```

```python
import math

class Diem:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Elip(Diem):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a
        self.b = b

    def dien_tich(self):
        return math.pi * self.a * self.b

class DuongTron(Elip):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
```
